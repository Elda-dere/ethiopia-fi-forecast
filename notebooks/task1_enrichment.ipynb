## Step 3.2 – Explore the dataset structure
We check the distribution of record types, pillars, and indicators.

```python
import pandas as pd
import os

# Check if files exist
print("Files in data/raw:", os.listdir('../data/raw'))

# Load and inspect the unified dataset
df = pd.read_csv('../data/raw/ethiopia_fi_unified_data.csv')
print("\n✅ Loaded successfully. Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nFirst 2 rows:\n", df.head(2))
print("\nRecord types:", df['record_type'].value_counts())
```

```python
# 1. Basic overview
print("=== DATASET OVERVIEW ===")
print(f"Total rows: {len(df)}")
print(f"Total columns: {len(df.columns)}")
print("\nRecord types:")
print(df['record_type'].value_counts())
print("\nPillars:")
print(df['pillar'].value_counts())
print("\nConfidence levels:")
print(df['confidence'].value_counts())
```

```python
# 2. Temporal coverage – which years have data for which indicators?
obs = df[df['record_type'] == 'observation'].copy()
obs['year'] = pd.to_datetime(obs['observation_date']).dt.year
print("\nObservation years by indicator:")
print(obs.groupby('indicator_code')['year'].describe())
```

```python
# 3. List all unique indicators
print("\nAll indicators with counts:")
print(obs['indicator_code'].value_counts())
```

```python
# 4. Existing events
events = df[df['record_type'] == 'event']
print("\nExisting events:")
print(events[['record_id', 'category', 'observation_date']])
```

```python
# 5. Existing impact links
impacts = df[df['record_type'] == 'impact_link']
print(f"\nExisting impact links: {len(impacts)}")
if len(impacts) > 0:
    print(impacts[['related_indicator', 'impact_direction', 'impact_magnitude', 'lag_months']])
```
